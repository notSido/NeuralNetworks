```python
# Install and import requirements
%pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

# Define the embeddings dictionary
embeddings = {
    "dog": [0.5, 0.3, -0.2],
    "cat": [0.6, 0.3, -0.3],
    "car": [-0.1, 0.6, 0.3],
    "bike": [-0.2, 0.7, 0.5]
}

# Extract words and embeddings
words = list(embeddings.keys())
vectors = np.array([embeddings[word] for word in words])

# Plot the 3D visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Set viewpoint
ax.view_init(elev=10, azim=55)

# Plot origin (absolute zero)
ax.scatter(0, 0, 0, color='black', s=100, label='Origin')

# Define colors for each vector for visual distinction
colors = ['blue', 'green', 'red', 'purple']

# Calculate the arrow length ratio for better visibility
max_vector_length = np.max(np.linalg.norm(vectors, axis=1))
arrow_length_ratio = 0.1 / max_vector_length  # Smaller ratio to fit the vectors in the plot

for word, vector, color in zip(words, vectors, colors):
    # Plot word embedding as text with an offset for clarity
    ax.text(vector[0]*1.15, vector[1]*1.15, vector[2]*1.15, word, size=20)
    # Plot arrow from origin to embedding
    ax.quiver(0, 0, 0, vector[0], vector[1], vector[2], color=color, arrow_length_ratio=arrow_length_ratio, linewidth=1)

# Set the aspect of the plot to be equal for all axes
ax.set_aspect('auto')

# Setting the bounds for the axes to be slightly larger than the max absolute value
max_abs_value = np.abs(vectors).max() * 1.15
ax.set_xlim([-max_abs_value, max_abs_value])
ax.set_ylim([-max_abs_value, max_abs_value])
ax.set_zlim([-max_abs_value, max_abs_value])

# Label the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
